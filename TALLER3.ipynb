###Ejercicio 1: Dado A = np.arange(1, 26).reshape(5, 5)
###a) Extrae la sub-matriz de las 3 últimas filas y las 3 primeras columnas.
###b) Extrae la segunda columna completa.
###c) Extrae los elementos de los ángulos (esquinas) de la matriz.

```python
import numpy as np

A = np.arange(1, 26).reshape(5, 5)
print(A)
```

###a) Extrae la sub-matriz de las 3 últimas filas y las 3 primeras columnas.


```python
sub = A[-3:, :3]
print(sub)
```

###b) Extrae la segunda columna completa

```python
col = A[:, 1]
print(col)
```

###c) Extrae los elementos de los ángulos (esquinas) de la matriz.

```python
corners = [A[0,0], A[0,-1], A[-1,0], A[-1,-1]]
print(corners)
```

###Ejercicio 2: Con la misma matriz A:
###a) Crea una máscara booleana para los valores mayores a 12.
###b) Reemplaza esos valores por -1 en una copia del array.
###c) Usa np.where para reemplazar impares por 0 y mantener los pares.

```python
mask = A > 12
print(mask)
```

###b) Reemplaza esos valores por -1 en una copia del array.

```python
B = A.copy()
B[mask] = -1
print(B)
```

###c) Usa np.where para reemplazar impares por 0 y mantener los pares.

```python
C = np.where(A % 2 == 0, A, 0)
print(C)
```

###Ejercicio 3: Crea un vector de 24 elementos con arange(24).
###Reorganízalo en: (a) matriz 6×4, (b) tensor 2×3×4, (c) vector fila 1×24

###Ejercicio 4: Demuestra la diferencia vista vs. copia: modifica un elemento
###en un slice y verifica que el array original también cambia.

```python
A = np.arange(1, 10).reshape(3, 3)

slice_view = A[:2, :2]  # esto es una VISTA
slice_view[0, 0] = 999

print("Slice modificado:")
print(slice_view)

print("Array original:")
print(A)
```
