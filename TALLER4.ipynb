###Ejercicio 1: Crea una matriz 5×4 de enteros aleatorios entre 1 y 100 (seed=7).
###a) Calcula la media, máximo y mínimo de cada columna.
###b) Calcula la suma acumulada de la primera fila.
###c) Normaliza la matriz para que cada columna tenga valores entre 0 y 1.
###Fórmula: (x - min_col) / (max_col - min_col)

```python
import numpy as np

np.random.seed(7)
matriz = np.random.randint(1, 101, (5, 4))

print("Matriz:\n", matriz)

# a) Media, máximo y mínimo por columna
media = matriz.mean(axis=0)
maximo = matriz.max(axis=0)
minimo = matriz.min(axis=0)

print("\nMedia por columna:", media)
print("Máximo por columna:", maximo)
print("Mínimo por columna:", minimo)

# b) Suma acumulada de la primera fila
suma_acum = np.cumsum(matriz[0])
print("\nSuma acumulada (primera fila):", suma_acum)

# c) Normalización por columna
matriz_norm = (matriz - minimo) / (maximo - minimo)
print("\nMatriz normalizada:\n", matriz_norm)
```

###Ejercicio 2: Usando broadcasting:
###a) Crea un vector fila [1,2,3,4] y un vector columna [[10],[20],[30]].
###b) Suma ambos vectores y explica el shape del resultado.

```python
# a)
fila = np.array([1, 2, 3, 4])        # shape (4,)
columna = np.array([[10], [20], [30]])  # shape (3,1)

# b)
resultado = fila + columna
print("\nResultado:\n", resultado)
print("Shape del resultado:", resultado.shape)
```

###Ejercicio 3: Con el array de notas y nombres del ejemplo de argsort:
###a) ¿Cuántos estudiantes tienen nota mayor al promedio? (usa máscara booleana)
###b) Imprime solo los nombres de esos estudiantes.

```python
nombres = np.array(["Ana", "Luis", "Carlos", "Marta", "Sofía"])
notas = np.array([85, 70, 90, 60, 88])

# Promedio
promedio = notas.mean()

# a) máscara booleana
mask = notas > promedio
cantidad = np.sum(mask)

print("\nPromedio:", promedio)
print("Cantidad de estudiantes sobre el promedio:", cantidad)

# b) nombres filtrados
print("Estudiantes sobre el promedio:", nombres[mask])
```

###Ejercicio 4: Crea dos matrices 3×4 y concaténalas vertical y horizontalmente.
###Luego divide el resultado vertical en 3 partes iguales con vsplit.

```python
A = np.random.randint(1, 10, (3, 4))
B = np.random.randint(1, 10, (3, 4))

print("\nMatriz A:\n", A)
print("\nMatriz B:\n", B)

# Concatenación
vertical = np.vstack((A, B))
horizontal = np.hstack((A, B))

print("\nConcatenación vertical:\n", vertical)
print("\nConcatenación horizontal:\n", horizontal)

# División en 3 partes iguales
dividido = np.vsplit(vertical, 3)

print("\nDivisión en 3 partes:")
for i, parte in enumerate(dividido):
    print(f"Parte {i+1}:\n", parte)
```
