
# Ejercicio 1: Crea una lista con los múltiplos de 7 entre 1 y 100 usando comprehension.



```python
multiplos_de_7 = [n for n in range(1, 101) if n % 7 == 0]

print(multiplos_de_7)
```

#Ejercicio 2: Dado el diccionario datos = {'a':10, 'b':25, 'c':7, 'd':18}

# a)Crea una lista con las claves cuyo valor sea mayor a 12.
# b) Crea un nuevo diccionario con esas claves y sus valores al cuadrado.

```python
datos = {'a': 10, 'b': 25, 'c': 7, 'd': 18}

# a) Lista con las claves cuyo valor es mayor a 12
claves_mayores = [k for k, v in datos.items() if v > 12]

# b) Nuevo diccionario con valores al cuadrado (solo para esas claves)
dict_cuadrados = {k: v**2 for k, v in datos.items() if v > 12}

print(f"Claves mayores a 12: {claves_mayores}")
print(f"Diccionario transformado: {dict_cuadrados}")
```

#Ejercicio 3: Define una función 'resumen(lista)' que retorne un diccionario con: longitud, suma, promedio, mínimo y máximo de la lista.

```python
def resumen(lista):
    if not lista:  # Verificamos si la lista está vacía para evitar errores
        return "La lista está vacía"

    return {
        "longitud": len(lista),
        "suma": sum(lista),
        "promedio": sum(lista) / len(lista),
        "minimo": min(lista),
        "maximo": max(lista)
    }

# Prueba de la función
mi_lista = [10, 20, 30, 40, 50]
resultado = resumen(mi_lista)

print(resultado)
```

#Ejercicio 4: Usa map() para aplicar la función potencia(x, 3) a la lista [1,2,3,4,5].


```python
# 1. Definimos la función potencia
def potencia(x, exponente=3):
    return x ** exponente

# 2. Definimos la lista original
numeros = [1, 2, 3, 4, 5]

# 3. Aplicamos map()
# Nota: map devuelve un objeto iterador, por lo que lo convertimos a lista para ver el resultado
resultado = list(map(lambda x: potencia(x), numeros))

print(resultado)
```
